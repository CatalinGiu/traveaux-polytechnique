```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display, HTML
pd.set_option('display.max_rows', 1000)

results = pd.read_csv("results.csv")
```

```python
table = results.groupby(['algo', 'taille', 'serie']).mean()[['temps', 'r']].unstack('serie')
table
```

```python
table = results.groupby(['algo', 'taille', 'serie']).max()[['temps','r']].unstack('serie')
table
```

```python
table = results.groupby(['algo', 'taille', 'serie']).min()[['temps','r']].unstack('serie')
table
```
