```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display, HTML
pd.set_option('display.max_rows', 1000)
```

```python
results = pd.read_csv("results.csv").replace(-1, np.nan)
group_set_1 = results[results['set_type'] == 1].groupby(['algo','taille']).mean()[['temps']]
group_set_2 = results[results['set_type'] == 2].groupby(['algo','taille']).mean()[['temps']]
group_set_3 = results[results['set_type'] == 3].groupby(['algo','taille']).mean()[['temps']]
```

```python
def plot_df(df, name):
    ax = df.plot(figsize=(20,10), logy=True, logx=True, grid=True, fontsize=14, title=name)
    ax.set_ylabel('temps(ms)', fontsize=14);
    ax.set_xlabel('taille', fontsize=14);

plot_df(group_set_1.unstack('algo').loc[:, 'temps'], "Temps en fonction de la taille pour les echantillons 0-9")
plot_df(group_set_2.unstack('algo').loc[:, 'temps'], "Temps en fonction de la taille pour les echantillons 10-19")
plot_df(group_set_3.unstack('algo').loc[:, 'temps'], "Temps en fonction de la taille pour les echantillons 20-29")
```
